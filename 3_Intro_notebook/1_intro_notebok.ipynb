## İklim değişikliği hakkında analiz
* geçmiş yılların analizi
* numpy ile **linear regression**
* `sklearn` ile regression

### Kaynakları okuyalım.

```python
import numpy as np
```

```python
yearsBase, meanBase = np.loadtxt('iklim/5-year-mean-1951-1980.csv', delimiter=',',usecols=(0, 1), unpack=True)
years, mean = np.loadtxt('iklim/5-year-mean-1882-2014.csv', delimiter=',',usecols=(0, 1), unpack=True)
```

```python
#list(zip(yearsBase,meanBase))
#list(zip(years,mean))
print(min(yearsBase),max(yearsBase))
print(min(years),max(years))
```

### Görselleştirelim

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(yearsBase, meanBase)

plt.title('scatter plot of mean temp difference vs year')
plt.xlabel('years', fontsize=12)
plt.ylabel('mean temp difference', fontsize=12)
plt.show()
```

### Noktaları ifade eden doğru denklemini bulalım.

```python
m,b=np.polyfit(yearsBase,meanBase,1)
```

```python
#Doğru denkleminin fonksiyonu oluşturalım
def f(x):
    return m*x+b
```

```python
#doğru denklemini alalım
print(' y = {0} * x + {1}'.format(m, b))
```

```python
#Doğru denklemini çizelim
plt.scatter(yearsBase, meanBase)
plt.plot(yearsBase, f(yearsBase))

plt.title('scatter plot of mean temp difference vs year')
plt.xlabel('years', fontsize=12)
plt.ylabel('mean temp difference', fontsize=12)
plt.show()
```

### Doğru denklemini lenear regression ile bulalım.

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True)

#modeli eğitelim
model.fit(yearsBase[:, np.newaxis], meanBase) #np.newaxis null anlamında

print(' y = {0} * x + {1}'.format(model.coef_[0], model.intercept_))
```

```python
mean_predicted = model.predict(yearsBase[:, np.newaxis])

plt.scatter(yearsBase, meanBase)
plt.plot(yearsBase, mean_predicted)

plt.title('scatter plot of mean temp difference vs year')
plt.xlabel('years', fontsize=12)
plt.ylabel('mean temp difference', fontsize=12)
plt.show()
```
